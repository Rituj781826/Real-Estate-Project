```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import math
from google.colab import drive
drive.mount('/content/drive')

# Load the dataset
flats= pd.read_csv("/content/drive/My Drive/flats.csv")

# Convert 'price' column to numeric
flats['price'] = pd.to_numeric(flats['price']
                               .str.replace(',', '')
                               .str.replace(' Lac', 'e5')  # Lac to INR
                               .str.replace(' Crore', 'e7') # Crore to INR
                               .str.replace('₹ ', ''),
                               errors='coerce')
# Convert 'area' column to numeric
flats['area'] = pd.to_numeric(flats['area'].str.replace(',', '').str.replace('₹ ', '').str.replace('/sq.ft.', '').str.extract(r'(\d+)')[0], errors='coerce')

# Convert 'areaWithType' to numeric
flats['area_sq_ft'] = pd.to_numeric(flats['areaWithType'].str.replace('Carpet area: ', '').str.replace('Built Up area: ', '').str.replace('Super Built up area ', '').str.extract(r'(\d+)')[0], errors='coerce')

# Convert 'bedRoom' and 'bathroom' to numeric
flats['bedRoom'] = pd.to_numeric(flats['bedRoom'].str.extract(r'(\d+)')[0], errors='coerce')
flats['bathroom'] = pd.to_numeric(flats['bathroom'].str.extract(r'(\d+)')[0], errors='coerce')

#Fill NaN values with 0 (or any other desired value)
flats = flats.fillna(0)

# Rest of the code remains the same

# Rest of the code remains the same

# Price Distribution
plt.figure(figsize=(10, 8))
sns.histplot(data=flats, x='price', kde=True, bins=30)
plt.title('Price Distribution')
plt.xlabel('Price (INR)')
plt.ylabel('Count')
plt.show()
#Area distribution
plt.figure(figsize=(10, 8))
sns.histplot(data=flats, x='area', kde=True, bins=30)
plt.title('Area Distribution')
plt.xlabel('Area(sq.ft.')
plt.ylabel('Count')
plt.show()


# Area vs. Price Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(flats['area'], flats['price'])
plt.title('Area vs. Price')
plt.xlabel('Area (sq. ft.)')
plt.ylabel('Price (INR)')
plt.show()

# Bedrooms and Bathrooms Bar Plot
plt.figure(figsize=(10, 6))
sns.countplot(data=flats, x='bedRoom')
plt.title('Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=flats, x='bathroom')
plt.title('Number of Bathrooms')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Count')
plt.show()

# Scatter Plot: Bedrooms vs. Price
plt.figure(figsize=(10, 6))
plt.scatter(flats['bedRoom'], flats['price'])
plt.title('Number of Bedrooms vs. Price')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price (INR)')
plt.show()

# Bubble Chart: Bathrooms vs. Price
plt.figure(figsize=(10, 6))
plt.scatter(flats['bathroom'], flats['price'], s=flats['area_sq_ft']*0.01, alpha=0.5)
plt.title('Number of Bathrooms vs. Price')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Price (INR)')
plt.show()
# Age and Possession Analysis
# Convert 'agePossession' to numeric representing the age in years
def convert_age(year_string):
    if re.match(r'^\d{4}$', year_string):  # Format: YYYY
        return abs(2024 - int(year_string))
    elif re.match(r'^[A-Za-z]{3} \d{4}$', year_string):  # Format: Mon YYYY
        month, year = year_string.split()
        return abs((2024 - int(year))) * 12  # Convert years to months
    elif re.match(r'^\d+ to \d+ Year Old$', year_string):  # Format: X to Y Year Old
        age_range = re.findall(r'\d+', year_string)
        return (int(age_range[0]) + int(age_range[1])) / 2
    elif year_string == 'Under Construction':
        return None  # Handle this case as per your requirement
    elif year_string == 'undefined':
        return None  # Handle this case as per your requirement
    else:
        return None  # Handle other cases as needed

# Apply the conversion function to the 'agePossession' column
flats['agePossession'] = flats['agePossession'].astype(str).apply(convert_age)

# Plot the histogram or density plot of property ages
plt.figure(figsize=(10, 6))
sns.histplot(data=flats, x='agePossession', kde=True, bins=30)
plt.title('Property Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Count')
plt.show()
'''Price Distribution:
The histogram visualizes the distribution of prices for a set of properties or items. The x-axis represents the price range, while the y-axis shows the count or frequency of items at each price level.
The distribution appears to be right-skewed, with the majority of the items concentrated towards the lower price range. The highest bar represents the most frequent price level, with a significant number of items falling within that price bin. As the prices increase, the frequency of items gradually decreases, forming a long tail on the right side of the distribution.
Area distribution:
it is a histogram showing the distribution of areas (in square feet) for a set of properties or houses. The x-axis represents the area range, while the y-axis shows the frequency or count of properties at each area bin. The distribution appears to be heavily right-skewed, with the majority of properties having smaller areas concentrated towards the left side of the plot. There is a sharp peak at a specific area range, indicating a high frequency of properties within that bin.
Area vs. Price Scatter Plot:
This scatter plot shows the relationship between the area (x-axis) and price (y-axis) of properties. Each data point represents an individual property, with its area and corresponding price plotted. The scatter plot reveals a positive correlation, where larger areas generally correspond to higher prices. However, there is a significant spread in the data points, suggesting that factors other than area also influence the price.
Number of Bedrooms Bar Chart:
This bar chart displays the distribution of the number of bedrooms in the properties. The x-axis represents the number of bedrooms, and the y-axis shows the count or frequency of properties with that number of bedrooms. The chart indicates that the majority of properties have either 2 or 3 bedrooms, with smaller numbers of properties having fewer or more bedrooms.
 Number of Bathrooms Bar Chart:
This bar chart displays the distribution of the number of bathrooms in the properties. The x-axis represents the number of bathrooms, and the y-axis shows the count or frequency of properties with that number of bathrooms. The chart indicates that most properties have either 1 or 2 bathrooms, with fewer properties having more bathrooms.
  Image 1:
The first plot shows the relationship between the number of bed narooms and the property price. It appears that as the number of bedrooms increases, the price tends to increase as well, though there is some variation in the data points.

The second plot displays the relationship between the number of bathrooms and the property price. Similarly, it suggests that properties with more bathrooms tend to have higher prices, but again, there is variability in the data.

The third plot is a histogram showing the age distribution of the properties. It indicates that most properties in the dataset are relatively new, with the majority being around 0-10 years old.
 Image 4:
The top plot is a correlation matrix displaying the correlations between different numerical variables in the dataset, such as price, area, number of bedrooms, number of bathrooms, and age of possession. The darker red shades indicate stronger positive correlations, while the darker blue shades represent stronger negative correlations.

The bottom set of plots contains various visualizations, including histograms and scatterplots, for the numerical variables in the dataset, allowing for an exploration of their distributions and relationships.'''
```
